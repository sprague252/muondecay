```python
from multiprocessing import Process, Value
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def detect(detecting):
    keepgoing = detecting.value
    while keepgoing:
        print('Detecting ...')
        time.sleep(5)
        keepgoing = detecting.value
    print('Stopping ...')
    if detecting:
        print('detecting evaluates to true')
    else:
        print('detecting evaluates to false')
    print('keepgoing: ', keepgoing)
    

if __name__ == '__main__':
    detecting = Value('i', 1)
    p = Process(target=detect, args=(detecting,))
    p.start()
    time.sleep(19)
    print('Time to stop!')
    detecting.value = 0
    p.join()
    print('Back together...All done')
```
