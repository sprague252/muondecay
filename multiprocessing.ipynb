```python
from multiprocessing import Process, Value
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def detect(detecting):
    while detecting.value:
        print('Detecting ...')
        time.sleep(3)
    print('Stopping ...')
    

if __name__ == '__main__':
    detecting = Value('i', 1)
    p = Process(target=detect, args=(detecting,))
    p.start()
    time.sleep(29)
    print('Time to stop!')
    detecting.value = 0
    p.join()
```
